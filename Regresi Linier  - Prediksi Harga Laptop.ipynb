Menghapus kolom yang tidak akan digunakan

```python
df.drop(columns=['LaptopID','ScreenResolution','Cpu','Gpu','OpSys','Weight'],inplace=True)
```

```python
df
```

```python
import re
df['Memory'] = df['Memory'].apply(lambda string: re.findall(r'\d+',str(string))[0])
df
```

```python
df['Ram'] = df['Ram'].str.replace('GB','')
df
```

```python
df.drop(columns=['Company','TypeName'],inplace=True)
```

Import Library

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Price']),df['Price'],test_size=0.2)
```

**Pembuatan Model Regresi Linier**

```python
mdl = LinearRegression()
mdl.fit(X_train,y_train)
```

Uji Normalitas

```python
from scipy import stats
jb_statistic, jb_p_value = stats.jarque_bera(df['Price'].values.reshape((-1,1)))

print("Nilai p- : ", jb_p_value)
```

**Evaluasi Linier**

Interpretasi Nilai MSE

```python
from sklearn.metrics import mean_squared_error
y_pred = mdl.predict(X_test[['Inches','Ram','Memory']])
mse = mean_squared_error(y_test, y_pred)
print("Nilai MSE:", mse)
```

Interpretasi Nilai MAE

```python
from sklearn.metrics import mean_absolute_error
y_pred = mdl.predict(X_test[['Inches','Ram','Memory']])
mae = mean_absolute_error(y_test, y_pred)
print("Nilai MAE:", mae)
```

Interpretasi Nilai RMSE

```python
from sklearn.metrics import mean_squared_error
y_pred = mdl.predict(X_test[['Inches','Ram','Memory']])
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Nilai RMSE:", rmse)
```

Interpretasi Nilai MAPE

```python
from sklearn.metrics import mean_absolute_percentage_error
y_pred = mdl.predict(X_test[['Inches','Ram','Memory']])
mape = mean_absolute_percentage_error(y_test, y_pred)

print("Nilai MAPE:", mape)
```

Hasil Prediksi

```python
score = mdl.score(X_train,y_train)
print("Nilai Keakuratan Prediksi : ", score)
```
